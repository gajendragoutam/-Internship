```python
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
from sklearn.model_selection import train_test_split , cross_val_score , GridSearchCV
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
glass.columns
```

```python
feature= glass[['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']]
zscore= np.abs(zscore(feature))
o_outlier= glass[(zscore < 3).all(axis=1)]
```

```python
print(f"Before",glass.shape)
print(f'After', o_outlier.shape)
```

```python
x= o_outlier.drop("Type", axis = 1)
y= o_outlier["Type"]
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
scale = StandardScaler()
```

```python
X_train= scale.fit_transform(x_train)
X_test = scale.transform(x_test)
```

```python
X_train
```

```python
LR = LogisticRegression()
LR.fit(X_train, y_train)
pred_LR= LR.predict(X_test)
pred_train= LR.predict(X_train)
score_LR=LR.score(X_test, y_test)
train_accuracy = LR.score(X_train, y_train)

print(LR)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_LR))
print('Classification Report', classification_report(y_test, pred_LR))
print(' Accuracy  score:', accuracy_score(y_test, pred_LR))
print('Accuracy on training data is ' , train_accuracy)
```

```python
RF = RandomForestClassifier ()
RF.fit(X_train, y_train)
pred_RF= RF.predict(X_test)
pred_train= RF.predict(X_train)
score_RF= RF.score(X_test, y_test)
train_accuracy = RF.score(X_train, y_train)

print(RF)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_RF))
print('Classification Report', classification_report(y_test, pred_RF))
print(' Accuracy  score:', accuracy_score(y_test, pred_RF))
print('Accuracy on training data is ' , train_accuracy)
```

```python
GB = GradientBoostingClassifier ()
GB.fit(X_train, y_train)
pred_GB= GB.predict(X_test)
pred_train= GB.predict(X_train)
score_GB= GB.score(X_test, y_test)
train_accuracy = GB.score(X_train, y_train)

print(GB)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_GB))
print('Classification Report', classification_report(y_test, pred_GB))
print(' Accuracy  score:', accuracy_score(y_test, pred_GB))
print('Accuracy on training data is ' , train_accuracy)
```

```python
DT = DecisionTreeClassifier ()
DT.fit(X_train, y_train)
pred_DT1 = DT.predict(X_test)
pred_train= DT.predict(X_train)
score_DT = DT.score(X_test, y_test)
train_accuracy1 = DT.score(X_train, y_train)

print(DT)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_DT))
print('Classification Report', classification_report(y_test, pred_DT))
print(' Accuracy  score:', accuracy_score(y_test, pred_DT1))
print('Accuracy on training data is ' , train_accuracy1)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)
pred_train = svm.predict(X_train)
score_svm = svm.score(X_test, y_test)  
train_accuracy = svm.score(X_train, y_train)  

print(svm)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_svm))
print('Classification Report', classification_report(y_test, pred_svm))
print(' Accuracy  score:', accuracy_score(y_test, pred_svm))
print('Accuracy on training data is ' , train_accuracy)
```

```python
KNN = KNeighborsClassifier ()
KNN.fit(X_train, y_train)
pred_KNN= KNN.predict(X_test)
pred_train= KNN.predict(X_train)
score_KNN= KNN.score(X_test, y_test)
train_accuracy = KNN.score(X_train, y_train)

print(KNN)
print()
print('Confusion Matrix', confusion_matrix(y_test, pred_KNN))
print('Classification Report', classification_report(y_test, pred_KNN))
print(' Accuracy  score:', accuracy_score(y_test, pred_KNN))
print('Accuracy on training data is ' , train_accuracy)
```

```python
score = cross_val_score(svm, X_train, y_train, cv=5)
print('Cross-Validation Scores:', score)
print('Mean Cross-Validation Score:', score.mean())
```

```python
score = cross_val_score(KNN , X_train, y_train, cv=5)
print('Cross-Validation Scores:', score)
print('Mean Cross-Validation Score:', score.mean())
```

```python
score = cross_val_score(GB, X_train, y_train, cv=5)
print('Cross-Validation Scores:', score)
print('Mean Cross-Validation Score:', score.mean())
```

```python
Altm = pd.DataFrame({
    'Algorithm' :  ['LogisticRegressor', ' DecisionTreeClassifier',
                    'GradientBoostingClassifier', 'RandomForestClassifier', 'SVC','KNeighborsClassifier' ],
                     
    'score' :  [score_LR, score_DT, score_GB, score_RF, score_svm , score_KNN]

})
Altm.sort_values(by='score', inplace=True)
```

```python
sns.barplot(x='Algorithm', y='score', data=Altm , palette= "Blues")
plt.xticks(rotation=45)
plt.title("Performance Comparison of Algorithms")
plt.show()
```

```python
parameter = {
    'C':             [0.1, 1, 10, 100],   
    'gamma':         ['scale', 'auto'],              
    'degree':        [2, 3, 4]                     
}

model = GridSearchCV(svm , parameter, cv=5 )
model.fit(X_train, y_train)
print("Best Parameters:", model.best_params_)
```

```python
best_model = model.best_estimator_
y_pred = best_model.predict(X_test)
print("Accuracy score " , accuracy_score(y_test, y_pred))
print("Confusion Matrix  " , confusion_matrix(y_test, y_pred))
print("Classification Report  ", classification_report(y_test, y_pred))
```
